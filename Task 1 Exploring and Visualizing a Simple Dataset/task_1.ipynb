```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris_df = sns.load_dataset('iris')
```

```python
iris_df.info()
```

```python
iris_df.describe()
```

```python
sns.pairplot(iris_df, hue='species')
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

sns.histplot(data=iris_df, x='sepal_length', hue='species', ax=axes[0, 0])
sns.histplot(data=iris_df, x='sepal_width', hue='species', ax=axes[0, 1])
sns.histplot(data=iris_df, x='petal_length', hue='species', ax=axes[1, 0])
sns.histplot(data=iris_df, x='petal_width', hue='species', ax=axes[1, 1])

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

sns.boxplot(data=iris_df, x='sepal_length', hue='species', ax=axes[0, 0])
sns.boxplot(data=iris_df, x='sepal_width', hue='species', ax=axes[0, 1])
sns.boxplot(data=iris_df, x='petal_length', hue='species', ax=axes[1, 0])
sns.boxplot(data=iris_df, x='petal_width', hue='species', ax=axes[1, 1])

plt.tight_layout()
plt.show()
```
